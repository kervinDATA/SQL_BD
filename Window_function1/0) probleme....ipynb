## Fenetrage: le principe

Pour l'instant, on a une table dans laquelle <br />
toutes les valeurs sont mélangées

```python
colors = {
    2:'background-color: #F6F7F3',
    3:'background-color: #124588',
    4:'background-color: #CDFFE4',
    5:'background-color: #A2D1FD',
    6:'background-color: #F7927E',
    7:'background-color: #C3322F',
}
   

def highlight_day(s):
    return [colors.get(s.loc['weekday']) for v in s.index]


print("la table actuelle:")
display(df.style.apply(highlight_day, axis=1))
```

Il faudrait comparer les Mardi avec les Mardi, <br />
Et les Samedi avec les Samedi

```python
(
    df
    .sort_values(["weekday", "date"])
    .head(20)
    .assign(
        count_lag=lambda df_: df_["visiteurs_count"].shift(1),
        rolling_mean=lambda df_: df_.groupby("weekday")["visiteurs_count"].rolling(4, min_periods=1, closed="left").mean().values
    )
    .style.apply(highlight_day, axis=1)
)
```

Dans ce chapitre, on va découvrir la notion de fenêtrage, <br />
qui permet de résoudre ce problème.
