# Agrégations

Agréger, c'est rassembler de nombreuses unités en une seule.

_Most rocks are aggregates of minerals_

Aggreger vient du Latin aggregare, qui veut dire "ajouter à", "rassembler en une masse ou un ensemble"

En maths, ça revient à prendre plusieurs valeurs et à n'en renvoyer qu'une seule.

```python
import pandas as pd
import duckdb
```

```python
notes = pd.DataFrame([[10, 20, 10, 0, 10, 20, 10, 0, 5, 15],
                      [10, 10, 10, 10, 10, 10, 10, 10, 10, 10]
                     ]).T
notes.columns = ["philo", "maths"]
notes
```

# Moyenne, écart type

Plusieurs valeurs, un seul résultat

```python
notes.mean()
```

```python
notes.std()
```

```python
query = """
SELECT AVG(philo), MEAN(maths) 
FROM notes
"""
duckdb.sql(query)
```
